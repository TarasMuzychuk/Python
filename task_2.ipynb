## Завдання № 1
Написати функцію з одним аргументом, яка перевіряє чи є аргумент парним чи непарним. Функція повинна повернути `even` або `odd` в залежності від того яке то число. Якщо аргумент `None` або не є числом -- необхідно повернути пусту строку.

```python
from __future__ import annotations

def is_even(arg: int) -> str:
 if (arg %  2 ==  0 ):  
    return "even"
 elif (arg %  2 != 0 ): 
    return "odd" 
 else:
     return ""
```

## Завдання № 2
Написати функцію, яка починає перебирати числа від 1 до нескінченності і перевіряє чи є число простим (просте число - те яке ділиться лише на 1 чи само на себе). Якщо просте число знайдене - воно додається до списку `numbers`. Після пʼятого знайденого простого числа функція повинна повернути суму всіх пʼяти простих чисел у списку.

```python
from __future__ import annotations

def sum_simple() -> int:
  d=30
  numbers = []
  for i in range(2, d+1):
        k = 0 
        for j in range(2, i+1):
         if i % j == 0 and i!=j and j!=1 :
           k +=1
        if k==0:
         numbers.append(i)
        else:
          k=0

  otvet = sum(numbers[:5])
  return otvet
 
print(sum_simple())
```

## Завдання № 3
Написати функцію з одним аргументом `n`, що є числом. Функція повинна повернути суму наступного ряду з `n` чисел $1 + 11 + 111 + 1111 + ....$, де кожний елемент - число з постійно зростаючою кількістю одиниць.

Наприклад:

Ви передаєте в функцію аргумент `5`.

Функція рахує сумму для ряду виду: $1 + 11 + 111 + 1111 + 11111$.

В даному прикладі вона повинна повернути число `12345`

```python
from __future__ import annotations

def sum_ones(n: int ) -> int:
    otvet = 0
    k = 1
    
    for i in range(1, n + 1):
     otvet = otvet + k         
     k = (k * 10) + 1     
         
    return otvet

n = 5
print(sum_ones(n))
```

В організації повинно бути 3 типи робітників `Employee`: `Developer`, `Designer` & `Manager`.

У кожного `Employee` повинні бути поля:

У `Designer` додатково повинен бути коефіцієнт ефективності `eff_coef`- число від 0 до 1, включно. Якщо коефіцієнт виходить за межі цього діапазону - необхідно поставити значення 0 (якщо він негативний) або 1 (якщо він більше 1).

У `Manager` повинна бути команда (список) що складається з довільної кількості `Developer` та `Designer`.

Створити обʼєкт `Department`, що містить список `Manager` (які в свою чергу містять команди).
Department повинен мати метод give_salary() який повинен надрукувати зарплату ВСІХ `Employee` цієї організації в форматі __іʼмя призвіще отримав counted_salary шекєлей__.

Зарплата `counted_salary` для кожного `Employee` обраховується з поля `base_salary` (базова заробітна плата) з наступними умовами:
- якщо його досвід роботи більше 2 років, він отримує `base_salary + 200`, але якщо його досвід більше 5 років - він отримує `base_salary * 1.2 + 500`.
- Додатково, на зп `Designer` впливає коефіцієнт ефективності `counted_salary * eff_coef`.
- Додатково, зп Manager розраховується з `counted_salary` плюс:
    - Якщо в його команді більше 5 людей - `counted_salary + 200`, але якщо в його команді більше 10 людей то `counted_salary + 300`.
    - Якщо в його команді більше половини `Developer` - його кінцева зп збільшується ще на `10%`.

Додаткова інформація:
- Для використання методів з суперкласу можна використати метод `super()`.
- Перевірити чи є об’єкт екземпляром класу можна за допомогою методу `isinstance`. Його перший агрументом - екземпляр класу, який ми перевіряємо, а другий - імʼя (_батьківського_) класу.
```python
from __future__ import annotations

class Employee:
    def __init__(self, first_name: str, last_name: str, base_salary: int, experience: int) -> None:
        self.first_name = first_name
        self.last_name = last_name
        self.base_salary = base_salary
        self.experience = experience 
    
    def counted_salary(self) -> float:
        if self.experience > 5:
            return self.base_salary * 1.2 + 500
        elif self.experience > 2:
            return self.base_salary + 200
        else:
            return self.base_salary

class Developer(Employee):
    def __init__(self, first_name: str, last_name: str, base_salary: int ,experience: int) -> None:
        super().__init__(first_name, last_name, base_salary, experience)

class Designer(Employee):
    def __init__(self, first_name: str, last_name: str, base_salary: int, experience: int, efficiency_coefficient: float) -> None:
        super().__init__(first_name, last_name, base_salary, experience)
        if efficiency_coefficient < 0:
            self.efficiency_coefficient = 0
        elif efficiency_coefficient > 1:
            self.efficiency_coefficient = 1
        else:
            self.efficiency_coefficient = efficiency_coefficient
    
    def counted_salary(self) -> float:
        salary = super().counted_salary()
        salary *= self.efficiency_coefficient
        return salary

class Manager(Employee):
    def __init__(self, first_name: str, last_name: str, base_salary: int, experience: int, team: list[Worker]) -> None:
        super().__init__(first_name, last_name, base_salary, experience)
        self.team = team
    
    def counted_salary(self) -> float:
        base_salary = super().counted_salary()

        if len(self.team) > 10:
            base_salary += 300
        elif len(self.team) > 5:
            base_salary += 200

        developers_count = sum(isinstance(emp, Developer) for emp in self.team)
        if developers_count > len(self.team) / 2:
            base_salary *= 1.1

        return base_salary

class Department:
     managers: list[Manager] = []

     def give_salary(self):
        total_salary = 0
        for manager in self.managers:
            total_salary += manager.counted_salary()
            print(f"{manager.first_name} {manager.last_name} отримав(ла) зарплату у розмірі {manager.counted_salary()} шекелів")
            for employee in manager.team:
                if hasattr(employee, 'counted_salary'):
                    total_salary += employee.counted_salary()
                    print(f"{employee.first_name} {employee.last_name} отримав(ла) зарплату у розмірі {employee.counted_salary()} шекелів")
        print(f"Загальна сума зарплати для всіх працівників: {total_salary} шекелів")

# Створюємо працівників
developer1 = Developer("Джон", "Сміт", 500, 3)
developer2 = Developer("Анна", "Коваль", 1150, 5)
developer3 = Developer("Олександр", "Шевченко", 400, 1)
developer4 = Developer("Марія", "Коваль", 1000, 8)
developer5 = Developer("Наталія", "Сидоренко", 900, 10)
developer6 = Developer("Ігор", "Білий", 2500, 4)
developer7 = Developer("Євген", "Кравчук", 2150, 6)
developer8 = Developer("Максим", "Козлов", 4200, 2)
developer9 = Developer("Злагода", "Моршинська", 2150, 6)
developer10 = Developer("Рута", "Рутина", 600, 2)
designer1 = Designer("Мері", "Джейн", 350, 6, 0.8)
designer2 = Designer("Петро", "Петров", 110, 2, 0.4)
designer3 = Designer("Андрій", "Бондар", 800, 10, 0.8)
designer4 = Designer("Іван", "Іванов", 1000, 1, 0.5)
designer5 = Designer("Віталій", "Семенов", 750, 7, 0.3)
designer6 = Designer("Олег", "Козлов", 150, 11, 0.02)
designer7 = Designer("Оксана", "Даниленко", 900, 4, 0.4)
designer8 = Designer("Юлія", "Петренко", 6500, 7, 0.9)
designer9 = Designer("Ксенія", "Батьківна", 1500, 3, 0.9)
designer10 = Designer("Тарас", "Музичук", 1700, 8, 0.9)

# Створюємо відділ і додаємо менеджерів
department = Department()
manager1 = Manager("Василь", "Іваненко", 900, 7, [developer1, developer2, developer3, developer4, developer5,developer6,designer5, designer6, designer7, designer8,designer9,designer10])
manager2 = Manager("Олег", "Петренко", 800, 2, [developer7, developer8, developer9,developer10, designer1, designer2, designer3, designer4])
department.managers.append(manager1)
department.managers.append(manager2)

# Додаємо працівників до менеджера
manager1.team.append(developer1)
manager1.team.append(developer2)
manager1.team.append(developer3)
manager1.team.append(developer4)
manager1.team.append(developer5)
manager1.team.append(developer6)

manager2.team.append(developer7)
manager2.team.append(developer8)
manager2.team.append(developer9)
manager2.team.append(developer10)
manager2.team.append(designer1)
manager2.team.append(designer2)
manager2.team.append(designer3)
manager2.team.append(designer4)

manager1.team.append(designer5)
manager1.team.append(designer6)
manager1.team.append(designer7)
manager1.team.append(designer8)
manager1.team.append(designer9)
manager1.team.append(designer10)

# Виплачуємо зарплату
department.give_salary()
```
